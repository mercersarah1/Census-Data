```python
import censusdis.data as ced
from censusdis.datasets import ACS5
from censusdis import states
import censusdis.geography as cgeo
import pandas as pd

# The variable I want to download.
MALES_25_to_29 = "B01001_001E"
FEMALES_30_to_34 = "B01001_036E"
INCOME_LESS_THAN_10k = "B19001_002E"
INCOME_150k_to_199k = "B19001_016E"

#
ced.download(
    dataset=ACS5,
    vintage=2023,
    download_variables=[MALES_25_to_29, FEMALES_30_to_34, INCOME_LESS_THAN_10k, INCOME_150k_to_199k],
    state=states.NY,
    state_legislative_district_lower_chamber='036'
)
```
